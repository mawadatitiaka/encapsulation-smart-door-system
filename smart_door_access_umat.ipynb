# Smart Door Access System for UMaT

## Objective
The purpose of this program is to demonstrate the concept of **encapsulation** in Object-Oriented Programming (OOP). Encapsulation protects sensitive data by making it private and allowing controlled access through properties.

## Class Description
A `Staff` class is created to represent a UMaT staff member. The class contains:

- **Public Attribute:** `s_name`
  - Stores the name of the staff member.
  - It can be accessed directly from outside the class.

- **Private Attribute:** `__access_code`
  - Stores the staff member's access code.
  - It is declared private using double underscores (`__`) to prevent direct access and modification.

## Methods and Properties

### 1. `@property access_code`
This property allows only authorized users to view the access code. The user is prompted to enter an administrator password. If the correct password is entered, the access code is displayed; otherwise, an "Access Denied" message is returned.

### 2. `@access_code.setter`
This property updates the access code after validating its length. The new access code must contain at least six characters. If the validation is successful, the access code is updated; otherwise, an error message is displayed.

### 3. `display_info()`
This method displays the staff member's information, specifically the staff name.

## How Encapsulation is Achieved

Encapsulation is implemented by:
- Declaring the access code as a private attribute (`__access_code`).
- Preventing direct access to the access code.
- Using the `@property` decorator to control how the access code is viewed.
- Using the `@access_code.setter` decorator to validate the access code before it is updated.

This ensures that sensitive information is protected and that only valid access codes are stored.

## Conclusion

The program successfully demonstrates encapsulation by protecting the access code from direct access and modification. The `@property` decorator provides a clean and Pythonic way of controlling access to private data while maintaining data integrity and improving the security of the Smart Door Access System.

```
class Staff:
    def __init__(self, s_name, access_code):
        self.s_name = s_name
        self.__access_code = access_code

    # Property to view the access code (authorized users only)
    @property
    def access_code(self):
        password = input("Enter admin password to view the access code: ")

        if password == "UMAT123":
            return self.__access_code
        else:
            return "Access Denied! Unauthorized user."

    # Property setter to update the access code after validation
    @access_code.setter
    def access_code(self, new_code):
        if len(new_code) >= 6:
            self.__access_code = new_code
            print("Access code updated successfully.")
        else:
            print("Error: Access code must be at least 6 characters long.")

    # Method to display staff information
    def display_info(self):
        print("===== STAFF INFORMATION =====")
        print("Staff Name:", self.s_name)
        print("=============================")


# ---------------- MAIN PROGRAM ----------------

staff1 = Staff("Mawada Naziru", "ABC12345")

# Display staff information
staff1.display_info()

# View access code
print("\nViewing Access Code:")
print(staff1.access_code)

# Update access code
print("Updating Access Code...")
staff1.access_code = input("Enter a new access code: ")

# View updated access code
print("Viewing Updated Access Code:")
print(staff1.access_code)
```
